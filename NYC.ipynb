```python
import pandas as pd
```

```python
csvFile = pd.read_csv('311_Service_Requests_for_2009.csv')
```

```python
csvFile.shape
```

```python
csvFile.fillna(0,inplace = True)
```

```python
csvFile[['City','Complaint Type']] #city with complaint
```

```python
frame = csvFile.groupby('Complaint Type')
```

```python
frame
```

```python
import operator
dict = {}
for key , data in frame:
    dict[key] = data.shape[0]
sorted_x = sorted(dict.items(), key=operator.itemgetter(1),reverse = True)
```

```python
type(sorted_x)
```

```python
fr = pd.DataFrame(sorted_x)
```

```python
fr[0].head(10) # top 10 complaints
```
